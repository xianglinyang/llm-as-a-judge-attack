# Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Rectangle
import seaborn as sns

# Set style for better looking plots
plt.style.use('seaborn-v0_8')
sns.set_palette("Set2")

# Beautiful color palette
beautiful_colors = {
    'before': '#E8998D',  # Soft coral
    'after': '#6BB6FF',   # Bright blue
    'improvement': '#7FB069',  # Fresh green
    'degradation': '#FFB347'   # Warm orange
}
```

```python
# Organize data with proper structure for the new layout
datasets = ['Alpaca', 'ArenaHard']
eval_types = ['Pointwise', 'Pairwise']
data_dict = {
    ('Pointwise', 'Alpaca'): pointwise_alpaca,
    ('Pointwise', 'ArenaHard'): pointwise_arenahard,
    ('Pairwise', 'Alpaca'): pairwise_alpaca,
    ('Pairwise', 'ArenaHard'): pairwise_arenahard
}

# Create a comprehensive bar chart with improved layout
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Define attack method labels for better readability
attack_labels = {
    'holistic_rewrite_one_shot': 'Holistic Rewrite',
    'holistic_rewrite': 'Iterative Rewrite',
    'random': 'Random',
    'ucb': 'UCB',
}
attacks = ["holistic_rewrite_one_shot", "holistic_rewrite", "random", "ucb"]
for row in range(2):
    for col in range(2):
        eval_type = eval_types[row]
        dataset = datasets[col]
        ax = axes[row, col]
        
        # Get data for this subplot
        data = data_dict[(eval_type, dataset)]
        
        # Prepare data for plotting
        before_scores = [data[attack]['before'] for attack in attacks]
        after_scores = [data[attack]['after'] for attack in attacks]
        
        # Create bar positions
        x_pos = np.arange(len(attacks))
        width = 0.35
        
        # Create bars with beautiful colors
        bars1 = ax.bar(x_pos - width/2, before_scores, width, label='Before Style Control', 
                       color=beautiful_colors['before'], alpha=0.85, edgecolor='white', linewidth=1.5)
        bars2 = ax.bar(x_pos + width/2, after_scores, width, label='After Style Control',
                       color=beautiful_colors['after'], alpha=0.85, edgecolor='white', linewidth=1.5)
        
        # Add value labels on bars
        for bar in bars1:
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{height:.3f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
        
        for bar in bars2:
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{height:.3f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
        
        # Set titles - datasets only on top row, evaluation types only on left column
        if row == 0:  # Top row - show dataset names
            ax.set_title(dataset, fontsize=16, fontweight='bold', pad=20)
        
        # Set y-axis labels - evaluation types only on left column
        if col == 0:  # Left column - show evaluation types
            ax.set_ylabel(eval_type, fontsize=14, fontweight='bold', rotation=90, 
                         va='center', ha='center', labelpad=20)
        
        # Set x-axis
        ax.set_xticks(x_pos)
        ax.set_xticklabels([attack_labels.get(attack, attack) for attack in attacks], 
                           rotation=45, ha='right', fontsize=11)
        
        # Add legend only to top-right subplot
        if row == 0 and col == 0:
            ax.legend(loc='upper left', frameon=True, fancybox=True, shadow=True, fontsize=11)
        
        # Style improvements
        ax.grid(True, alpha=0.3, axis='y', linestyle='--')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(0.5)
        ax.spines['bottom'].set_linewidth(0.5)
        
        # Set y-axis limits with some padding
        all_scores = before_scores + after_scores
        y_min, y_max = min(all_scores), max(all_scores)
        y_range = y_max - y_min
        ax.set_ylim(y_min - 0.1 * y_range, y_max + 0.15 * y_range)

# Add overall title
# fig.suptitle('Style Control Effects on Different Attack Methods', fontsize=18, fontweight='bold', y=0.95)

plt.tight_layout()
plt.subplots_adjust(top=0.9, left=0.1, right=0.95)
plt.savefig('/home/xianglin/git_space/llm-as-a-judge-attack/plots/style_control_v2_effects.pdf', dpi=300, bbox_inches='tight')
```

```python
# Alternative visualization: Show the difference (improvement/degradation) caused by style control
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for row in range(2):
    for col in range(2):
        eval_type = eval_types[row]
        dataset = datasets[col]
        ax = axes[row, col]
        
        # Get data for this subplot
        data = data_dict[(eval_type, dataset)]
        
        # Calculate differences (after - before)
        attacks = list(data.keys())
        differences = [data[attack]['after'] - data[attack]['before'] 
                       for attack in attacks]
        
        # Create colors based on positive/negative differences
        bar_colors = [beautiful_colors['improvement'] if diff > 0 else beautiful_colors['degradation'] 
                      for diff in differences]
        
        # Create bars
        bars = ax.bar(range(len(attacks)), differences, color=bar_colors, alpha=0.85, 
                      edgecolor='white', linewidth=1.5)
        
        # Add value labels on bars
        for i, (bar, diff) in enumerate(zip(bars, differences)):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., 
                    height + 0.001 if height >= 0 else height - 0.005,
                    f'{diff:+.4f}', ha='center', 
                    va='bottom' if height >= 0 else 'top', 
                    fontsize=10, fontweight='bold')
        
        # Add horizontal line at y=0
        ax.axhline(y=0, color='black', linestyle='-', linewidth=1, alpha=0.7)
        
        # Set titles - datasets only on top row, evaluation types only on left column
        if row == 0:  # Top row - show dataset names
            ax.set_title(dataset, fontsize=16, fontweight='bold', pad=20)
        
        # Set y-axis labels - evaluation types only on left column
        if col == 0:  # Left column - show evaluation types
            ax.set_ylabel(eval_type, fontsize=14, fontweight='bold', rotation=90, 
                         va='center', ha='center', labelpad=20)
        
        # Set x-axis
        ax.set_xticks(range(len(attacks)))
        ax.set_xticklabels([attack_labels.get(attack, attack) for attack in attacks], 
                           rotation=45, ha='right', fontsize=11)
        
        # Style improvements
        ax.grid(True, alpha=0.3, axis='y', linestyle='--')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_linewidth(0.5)
        ax.spines['bottom'].set_linewidth(0.5)
        
        # Set y-axis limits with some padding
        y_max = max(abs(min(differences)), abs(max(differences)))
        ax.set_ylim(-y_max * 1.2, y_max * 1.2)

# Add a legend for the colors
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=beautiful_colors['improvement'], label='Improvement (Positive Change)'),
                   Patch(facecolor=beautiful_colors['degradation'], label='Degradation (Negative Change)')]
fig.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, 0.02), 
           ncol=2, frameon=True, fancybox=True, shadow=True, fontsize=12)

# Add overall title
fig.suptitle('Score Changes After Style Control', 
             fontsize=18, fontweight='bold', y=0.95)

plt.tight_layout()
plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1, right=0.95)
plt.show()
```
